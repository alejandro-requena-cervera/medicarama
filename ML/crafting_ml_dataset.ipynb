# Plotear distribución del target

```python
df_plot = df_main.value_counts(subset='Target', ascending=True)
df_plot.plot(kind='barh', figsize=(15, 12), color=c_med)
plt.grid(axis='x', alpha=0.33);
```

##### Puntos a observar del target:

- **Complejidad**: Es un target con una alta variabilidad (hay 52 cursos == 52 categorías diferentes en el target) y, además, tiene una distribución muy descompensada.
- **Dificultad para generalizar**: El dataset ha pasado de tener +400.000 filas a tener algo más de 14.000 filas. Esto y que hay 52 categorías diferentes en el target van a dificultar que el modelo generalice correctamente.


Hay que poner mucho incapié en que exista el mejor balanceo posible en la distribución de los datos en el target al dividir el dataset (usar el parámetro `stratify`).
